```python
# Drop the Duplicated Columns
duplicated_columns = ["LOCATION","Pollutant","Measure","Flight type", "Frequency", "Source of emissions","SEASONALITY",
                      "Seasonality", "Time", "MEASURE", "POLLUTANT"]
for i in range(len(duplicated_columns)):
    df.drop(duplicated_columns[i], axis = 1 , inplace = True)
df
```

```python
df.rename(columns= {"Value":"Co2 Emissions"}, inplace = True)
df
```

```python
# Calculating top 10 Co2 Emissions by Time Period
# Group the data by the 'Time' column and sum the emissions
top_emissions_by_Time = df.groupby('TIME')['Co2 Emissions'].sum().reset_index()

# Sort the data by emissions in descending order
top_emissions_by_Time = top_emissions_by_Time.sort_values(by='Co2 Emissions', ascending=False)

# Take the top 10 rows
top_10_emissions_by_Time = top_emissions_by_Time.head(10)

print("Top 10 CO2 emissions by country:")
print(top_10_emissions_by_Time)
```

```python
# Calculating top 10 Co2 Emissions by country
# Group the data by the 'Country' column and sum the emissions
TopCo2EmissionsByCountry = df.groupby('Country')['Co2 Emissions'].sum().reset_index()

# Sort the data by emissions in descending order
TopCo2EmissionsByCountry = TopCo2EmissionsByCountry.sort_values(by='Co2 Emissions', ascending=False)

# Take the top 10 rows
Top10Co2EmissionsByCountry = TopCo2EmissionsByCountry.head(10)

print("Top 10 CO2 emissions by country:")
Top10Co2EmissionsByCountry
```

```python
list_of_top_ten_country = Top10Co2EmissionsByCountry["Country"].tolist()
print(list_of_top_ten_country)
```

```python
def IsItImpactOnClimateChange(Co2_Emissions, FLIGHT, Country):
    if Co2_Emissions > 100 or FLIGHT == "P" or Country in list_of_top_ten_country:
        return 1
    return 0
```

```python
#add a new Feature "Climate Change Impact"
list_climate_change_impact = []
for i in range(df.shape[0]):
    list_climate_change_impact.append(IsItImpactOnClimateChange(df["Co2 Emissions"][i], df["FLIGHT"][i], df["Country"][i]))
df = df.assign(Climate_Change_Impact = list_climate_change_impact)
df
```

```python
# Encode the categorical data to numerical first we will encode the country with label encoding
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

# Fit and transform Country data
df["Country"] = label_encoder.fit_transform(df["Country"])
df
```

```python
# Encode the categorical data to numerical will encode remaining features Except the date with Dummy Encoding
df = pd.get_dummies(df, columns=['FLIGHT', 'FREQUENCY', 'SOURCE'], drop_first=True, dtype= int)
df
```

```python
df.drop("TIME", axis =1 , inplace = True)
df
```

```python
from imblearn.over_sampling import SMOTE
X = df.drop("Climate_Change_Impact", axis = 1)
y = df["Climate_Change_Impact"]
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)
```

```python
df = pd.concat([X_resampled, pd.DataFrame(y_resampled, columns=['Climate_Change_Impact'])], axis=1)
df
```
